## ■ 파이썬으로 json 데이터 다루기

```python
# 필요한 패키지 임포트
import json

# __init__.py 파일이 있으면, 패키지(json)에서 모듈(file.py)를 거치지 않고도 바로 .찍고 호출 가능
```

```python
# json 문자열 데이터 생성 : 바깥쪽 '' + 안쪽 ""   <<<--- python과의 차이점

# python에서는 두가지 경우 모두 잘 작동
str1 = '{"name":"달러구트 꿈 백화점","author":"이미예"}'
str2 = "{'name':'달러구트 꿈 백화점','author':'이미예'}"

# 생성된 데이터 확인
print(f'python의 문자열 str1 : {str1}')
print(f'python의 문자열 str2 : {str2}')
```

```python
# json 문자열 --> python dict 자료형으로 변환 --> json.loads()
py_dict1 = json.loads(str1)
# py_dict2 = json.loads(str2)     # <-- 에러발생

# 변환된 결과 확인 (1)  --> python dict 확인 --> 자료형 확인
data_type = type(py_dict1)
print(f'변환된 결과 값의 자료형 확인 : {data_type}\n')

# 변환된 결과 확인 (2)  --> python dict --> 변수[key] --> value 추출
book_name = py_dict1['name']
print(f'검색의 결과 : py_dict1[\'name\'] = {book_name}')
```

```python
# API 호출을 통해서 획득한 json 문자열 데이터 --> python dict 자료형으로 변환

api_dict = json.loads(data)
print(f'변환결과 : \n{api_dict}\n')       # <--- 큰 따옴표가 작은 따옴표로 변환
print(f'변환결과 데이터 자료형 : type(api_dict) = {type(api_dict)}\n')
# python dict 자료형 --> 성분원소의 수 확인 --> len() 함수
print(f'성분원소의 수 : {len(api_dict)}\n')

# python dict 자료형 --> 변수.keys() --> key 추출]
dict_keys = api_dict.keys()
print(f'추출된 key를 확인 : {dict_keys}')

# python dict 자료형 --> 변수[key] --> value 추출
dict_values = api_dict['response']
print(f'추출된 value 확인 :\n{dict_values}\n')

# print(f'성분원소의 수 : {len(api_dict[0])}')  <-- 리스트가 아니기 때문에 인덱스로 지정하면 에러, 반드시 key로 지정해야 한다.
```

```python
# python dict --> json 문자열로 변환

# python : 문자열이 아닌 자료형 --> 문자열 변환  --> str() 함수 :::::: 이렇게 하면 안된다.
api_str = str(api_dict)
print(f'str() 함수를 이용해서 문자열로 변환한 결과 :\n{api_str}\n')       # <=== 안쪽이 작은 따옴표로 되어 있다.

# json.loads(api_str)     # <-- 실제로 해보면 에러 발생한다.


# json.dumps(dict, ensure_ascii=False)
# ensure_ascii=True (default) : 영어만 있을때 이렇게 사용, ensure_ascii=False(한글 사용시)
json_str = json.dumps(api_dict)
print(json_str)  # <== 한글이 다 깨진다.(16진수로 표현됨.)

# json.dumps(dict, ensure_ascii=False)
json_str = json.dumps(api_dict, ensure_ascii=False)
print(json_str)  # <== 이렇게 해야 됨
```

## ■ BeautifulSoup

```python
# 필요한 라이브러리 설치
# !pip install beautifulsoup4   <-- 이미 설치되어 있다.

# 필요한 클래스 함수 임포트
from bs4 import BeautifulSoup
```

```python
# 클래스 함수 호출결과 지정
###############################################################
# BeautifulSoup(html, 분석도구:parser)
###############################################################
# BeautifulSoup 분석도구의 종류
# 1. lxml        : 속도빠름, 별도의 라이브러리 설치 필요
# 2. lxml-xml    : 속도빠름, 별도의 라이브러리 설치 필요
# 3. html5lib    : 매우느림, 별도의 라이브러리 설치 필요
# 4. html.parser : 적당한 속도, 별도의 라이브러리 설치 불필요
###############################################################

bs = BeautifulSoup(html, 'lxml')

# 결과 확인하기
print(bs)
```

```python
# 상세 페이지로 연결되는 링크에 관한 정보 가져오기 --> 변수.select()
link = bs.select('a.gd_name')

# 결과 확인하기
print(link)
print(type(link))
# 리스트자료형은 아니지만 인덱싱이 가능하다.
```
