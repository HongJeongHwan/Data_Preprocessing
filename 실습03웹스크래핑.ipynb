## ■ 웹스크레핑을 이용한 네이버 뉴스 수집

```python
# 필요한 모듈, 함수 임포트
import requests
from bs4 import BeautifulSoup

# requests.get(url) 함수 이용 --> 네이버 언론사별 랭킹뉴스 웹페이지 html 수집
# 호출 url 작성
url = 'https://news.naver.com/main/ranking/popularDay.naver?mid=etc&sid1=111'
headers = {"useragent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"}

# 정보 요청 & 수령
response = requests.get(url,headers=headers)

# html 정보 확인
html= response.text
# print(html)

# BeautifulSoup 클래스 호출, 실행
bs = BeautifulSoup(html, 'lxml')

# 전체 언론사의 랭킹뉴스 정보 추출(tag.class) --> list로 반환
news_list = bs.select('div.rankingnews_box')

# 결과 확인하기
# print(f'성분원소의 개수 = {len(news_list)}\n')
# print(f'수집한 결과 확인 :\n{news_list}')


# 추출한 언론사의 랭킹뉴스 정보(html) --> 확인 --> 1개 언론사 정보
# news_list 변수 --> 82개 항목 --> index=0 추출
press = news_list[0]
# print(press)
# 언론사 이름(하나만 추출) : 지정한 태그의 html을 가져온다.
html_name = press.select_one('strong.rankingnews_name')
# print(html_name)
name = html_name.text
name = html_name.get_text()
# print(f'언론사의 이름 : {name}')

# 5개의 상위 랭킹뉴스 추출 --> html 추출
news5 = news_list[0].select('li')
print(news5)
# <li> 태그 --> 뉴스랭킹(list_ranking_num), 뉴스링크와 뉴스제목(list_title), 작성 시간, 썸네일(img) 정보 추출
for li in news5:
    # 뉴스 랭킹정보 추출
    news_ranking = li.select_one('em.list_ranking_num').get_text()
    print(f'뉴스 랭킹정보 : {news_ranking}')
    print('-'*80)
    # 뉴스 제목 추출
    news_title = li.select_one('a.list_title').get_text()
    print(f'뉴스 제목 : {news_title}')
    print('-'*80)
    # 뉴스 링크 추출
    news_link = li.select_one('a.list_title')['href']
    print(f'뉴스 링크 : {news_link}')

    print('='*80)

    # 결과 확인하기
    print(f'언론사 이름 : {name}')
    print(f'뉴스 랭킹정보 : {news_ranking}')
    print(f'뉴스 제목 : {news_title}')
    print(f'뉴스 링크 : {news_link}')

# press 내부에 들어있는 속성값 --> 변수[key] 방식으로 값 추출 : dict는 아니지만 dict같은 방식으로 값 추출 가능
# link_info = press['href']
```
